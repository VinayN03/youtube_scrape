```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import json
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
search_text = "portable table"
video_count = 200
```

```python
search = search_text.replace(' ', '+')
url = "https://www.youtube.com/results?search_query="+search_text+"&sp=CAMSAhAB"
# Define Chrome options to open the window in maximized mode
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")

# Initialize the Chrome webdriver and open the URL
driver = webdriver.Chrome(chrome_options=options)

driver.get(url)

body = driver.find_element_by_tag_name('body')
body.send_keys(Keys.END)
time.sleep(0.5)
    
# soup = BeautifulSoup(driver.page_source, "html.parser")


# with open('page_source.html', 'w+', errors='ignore', encoding = 'utf-8') as f:
#     f.write(page_source)
# print("page_source : ", page_source)
script_tags = driver.find_elements_by_tag_name('script')
script_tags = soup.findAll('script')
for i in script_tags:
    if 'window["ytInitialData"]' in str(i.get_attribute('innerHTML')):
        result_dict = json.loads(str(i.get_attribute('innerHTML')[31:-110]))

for k, v in result_dict.items():
    if k == 'contents':
        lst1 = v['twoColumnSearchResultsRenderer']['primaryContents']['sectionListRenderer']['contents']
        

        
video_lst = []
for i in lst1:
    for k1, v1 in i.items():
        if k1 == 'itemSectionRenderer':
            for j in v1['contents']:
                
                for k2, v2 in j.items():
                    if k2 == 'videoRenderer':
                        a_dict = {}
                        for k3, v3 in v2.items():
                            if k3 == 'videoId':
                                a_dict['Video ID'] = v3
                            elif k3 == 'title':
                                a_dict['Title'] = v3['runs'][0]['text']
                            elif k3 == 'publishedTimeText':
                                a_dict['Published Time'] = v3['simpleText']
                            elif k3 == 'lengthText':
                                a_dict['Video Length'] = v3['simpleText']
                            elif k3 == 'viewCountText':
                                a_dict['View Count'] = v3['simpleText']
                        video_lst.append(a_dict)
                        
driver.close()

df = pd.DataFrame(video_lst)
display(df)
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")

# Initialize the Chrome webdriver and open the URL
driver = webdriver.Chrome(chrome_options=options)

driver.get(url)
for n in range(2):
    body = driver.find_element_by_tag_name('body')
    body.send_keys(Keys.END)
    time.sleep(0.5)

soup = BeautifulSoup(driver.page_source, 'html.parser')
mydivs = soup.findAll("div", {"id":"dismissable", "class": "style-scope ytd-video-renderer"})
```

```python
video_lst = []
for i in mydivs:
#     print(i)
    a_dict = {}
    a_dict['Video Title'] = i.find("a", attrs={"id": "video-title"})['title']
    a_dict['Video ID'] = i.find("a", attrs={"id": "video-title"})['href'].replace('/watch?v=', '')
    a_dict['Video URL'] = 'https://youtube.com'+i.find("a", attrs={"id": "video-title"})['href']
    a_dict['Video Length'] = ' '.join(i.find("a", attrs={"id": "video-title"})['aria-label'].split()[-4:-2])
    a_dict['Channel Link'] = i.find("ytd-channel-name").find('a')['href']
    a_dict['Channel Name'] = i.find("ytd-channel-name").find('a').get_text()
    a_dict['View Count'] = i.find("a", attrs={"id": "video-title"})['aria-label'].split()[-2]
    a_dict['Description'] = i.find("yt-formatted-string", attrs={"id": "description-text"}).get_text()
    
#           <div id="overlays" class="style-scope ytd-thumbnail"><ytd-thumbnail-overlay-time-status-renderer class="style-scope ytd-thumbnail" overlay-style="DEFAULT"><!--css-build:shady--><yt-icon class="style-scope ytd-thumbnail-overlay-time-status-renderer" disable-upgrade="" hidden=""></yt-icon><span class="style-scope ytd-thumbnail-overlay-time-status-renderer" aria-label="35 seconds">
# 
#     print(a_dict)
    video_lst.append(a_dict)
```
